>1. Sesiones de doblaje <br>


**Enunciado del trabajo:**

Se precisa coordinar el doblaje de una película. Los actores del doblaje deben coincidir en las
tomas en las que sus personajes aparecen juntos en las diferentes tomas. Los actores de
doblaje cobran todos la misma cantidad por cada día que deben desplazarse hasta el estudio de
grabación independientemente del número de tomas que se graben. No es posible grabar más
de 6 tomas por día. El objetivo es planificar las sesiones por día de manera que el gasto por los
servicios de los actores de doblaje sea el menor posible.


#Modelo

**¿Cómo represento el espacio de soluciones?**

En este problema se encuentran varias soluciones:
1. Un diccionario que representa los dias de grabacion y las tomas que se van a grabar
2. Una lista donde estan los actores que tienen que presentarse cada jornada de rodaje y el coste final del rodaje.

**¿Cuál es la función objetivo?**

El fin es el coste total más optimo de la grabacion.

**¿Cómo implemento las restricciones?**

Para ello dl algoritmo que se implementan cumple las siguientes 3 restricciones:
1.   Restringe que las tomas se grabes siempre en un dia y no en dias separados.
2.   Restringe que el numero de tomas maximo por dia sean 6.
3.   Los actores que coinciden en las escenas se asignen el mismo dia.


#Análisis
**¿Que complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones**

Debido a que es un problema de asignacion la complejidad podria ser tanto exponencial como factorial. Pero las caracteristicas concretas de este problema nos hacen ver que finalmente la complejidad es exponencial del tipo O(m^n).

- n = numero de tomas
- m = numero de dias disponibles

- Sencilla de desarrollar.

- Aporta soluciones que se acercan a la solucion optima del problema

```python
## IMPORTAMOS LIBRERIAS
import numpy as np
import pandas as pd
from google.colab import files

```

```python
## CARGA Y VISUALIZACION DEL DATASET
# Se importa el archivo .csv
csv_file = files.upload()

# Paso necesario para poder leer el csv de forma correcta
dataframe = list(csv_file.keys())[0]

# Cargar, procesar y visualizar el dataframe
df_tomas_actores = pd.read_csv(dataframe, index_col=0)
# Se muestra el dataframe
print(df_tomas_actores.to_string())
```

```python
## PREPROCESADO DEL DATASET
df_tomas_actores_clean = df_tomas_actores.iloc[1:-3, :-2]
print('######################################[INFO]######################################\n')
print('Las dimensiones del dataset de tomas de los actores son las siguientes: ', df_tomas_actores_clean.shape)
print('\n##################################################################################')
print(df_tomas_actores_clean.to_string())
```

```python
## CONVERSION DEL DATASET A ARRAYS DE NUMPY
matriz_tomas_actores = df_tomas_actores_clean.values.tolist()
for fila in matriz_tomas_actores:
    print(fila)
```

```python
## PRIMER TIPO DE ALGORITMO: TÉCNICAS VORACES ##
# Inicializacion de las variables
horarios = {}
actores_presentes = []
jornada = 1
coste = 0

# Se calcula el orden de las tomas, en funcion del número de actores que participan en cada una
suma_filas = np.sum(matriz_tomas_actores, axis=1)
lista_tomas = sorted(range(len(matriz_tomas_actores)), key=suma_filas.__getitem__, reverse=True)

# Distribución de las tomas por dias de grabación
for toma in lista_tomas:
    toma_asignada = False
    for dia, tomas in horarios.items():
        if len(tomas) + 1 <= 6:
            toma_disponible = True
            for actor in range(len(matriz_tomas_actores[0])):
                if matriz_tomas_actores[toma][actor] and actor + 1 in tomas:
                    toma_disponible = False
                    break
            if toma_disponible:
                tomas.append(toma)
                toma_asignada = True
                break
    if not toma_asignada:
        horarios[len(horarios) + 1] = [toma]
```

```python
## PRESENTACION DE LOS RESULTADOS
print('######################################[INFO]######################################\n')
print('Los horarios calculados son los siguientes:')
for dia, tomas in horarios.items():
    print('El día', dia, 'se graban las tomas:', sorted(tomas))

print('\n-----------------------------------------------------------------------------------\n')
# Calcula el numero de actores presentes en cada dia de grabación
print('Los actores que se necesitan para grabar cada dia son los siguientes:')
for dia in horarios.values():
  actores_presentes = [actor + 1 for toma in dia for actor in range(len(matriz_tomas_actores[0])) if matriz_tomas_actores[toma][actor] == 1]
  actores_presentes_set = set(actores_presentes)
  print('El día', jornada, 'tienen que ir los actores:', actores_presentes_set)
  jornada = jornada + 1
print('\n-----------------------------------------------------------------------------------\n')
# coste = sum(actores_presentes_set for dia in horarios.values())
for actores_presentes_set in horarios.values():
    coste += len(set(actores_presentes_set))
print('El coste final es:', coste)
print('\n##################################################################################')
```

Para comprobar si puede haber mejores soluciones se prueba con un algoritmo de recocido simulado.

```python
## NUEVOS IMPORTS
import random
import math
from copy import deepcopy
```

```python
## FUNCIONES A EMPLEAR
# esta funcion genera los primeros valores de asignacion de manera aleatoria
def asignar_rodajes_aleatorios(rodajes):
    tomas = list(range(rodajes))
    random.shuffle(tomas)
    dias = [tomas[i:i+6] for i in range(0, rodajes, 6)]
    return dias
# esta funcion evalua los valores actuales considerados optimos frente a los nuevos valores
def evaluar_optimos(gasto, nuevo_gasto, temperatura):
    if nuevo_gasto < gasto:
      solucion = 1.0
    else:
      solucion = math.exp((gasto - nuevo_gasto) / temperatura)
    return solucion
# esta funcion calcula los costes
def calcular_coste(asignacion, matriz_tomas_actores):
    coste = sum(len(obtener_actores(dia, matriz_tomas_actores)) for dia in asignacion)
    return coste
# esta funcion obtiene los actores presentes el dia de la grabacion
def obtener_actores(dia, matriz_tomas_actores):
    actores_presentes = set()
    for rodaje in dia:
        actores_presentes.update(actor + 1 for actor, presencia in enumerate(matriz_tomas_actores[rodaje]) if presencia == 1)
    return actores_presentes
# esta funcion genera nuevas asignaciones
def generar_nueva_asignacion(asignacion):
    copia_asignacion = deepcopy(asignacion)
    dia1, dia2 = random.sample(range(len(asignacion)), 2)
    rodaje1 = random.choice(copia_asignacion[dia1])
    rodaje2 = random.choice(copia_asignacion[dia2])
    copia_asignacion[dia1].remove(rodaje1)
    copia_asignacion[dia2].remove(rodaje2)
    copia_asignacion[dia1].append(rodaje2)
    copia_asignacion[dia2].append(rodaje1)
    return copia_asignacion
```

```python
## SEGUNDO TIPO DE ALGORITMO: RECOCIDO SIMULADO ##
# Definicion de los parametros de inicio del algoritmo
temperatura = 10000 # temperatura inicial
temperatura_final = 0.00000000001
factor_enfriamiento = 0.2
iteraciones_por_temperatura = 1000
jornada = 1

# Comienzo del algoritmo de recocido, se hace una asignacion aleatoria de los rodajes
asignacion_actual = asignar_rodajes_aleatorios(len(matriz_tomas_actores))
# se toma el valor de la asignacion como el valor optimo
mejor_asignacion = asignacion_actual
# se calcula el coste inicial en base a la asignacion aleatoria
coste = calcular_coste(asignacion_actual, matriz_tomas_actores)
# se toma el valor del coste como el valor optimo
mejor_coste = coste

# inicia el bucle de recocido
while temperatura > temperatura_final:
    for _ in range(iteraciones_por_temperatura):
        # se generan nuevos valores de asignacion y de coste
        nueva_asignacion = generar_nueva_asignacion(asignacion_actual)
        nuevo_coste = calcular_coste(nueva_asignacion, matriz_tomas_actores)
        # se evalua si la ultima generacion de valores es la mejor
        if evaluar_optimos(coste, nuevo_coste, temperatura) > random.random():
            asignacion_actual = nueva_asignacion
            coste = nuevo_coste
        if nuevo_coste < mejor_coste:
            mejor_asignacion = nueva_asignacion
            mejor_coste = nuevo_coste
    # se actualiza el valor de la temperatura
    temperatura *= factor_enfriamiento
```

```python
## PRESENTACION DE LOS RESULTADOS
print('######################################[INFO]######################################\n')
print('Los horarios calculados son los siguientes:')
for dia, tomas in enumerate(mejor_asignacion):
    print('El día', dia, 'se graban las tomas:', sorted(tomas))

print('\n-----------------------------------------------------------------------------------\n')
# Calcula el numero de actores presentes en cada dia de grabación
print('Los actores que se necesitan para grabar cada dia son los siguientes:')
for i, dia in enumerate(mejor_asignacion):
    actores_presentes = obtener_actores(dia, matriz_tomas_actores)
    print('El día', jornada, 'tienen que ir los actores:', actores_presentes)
    jornada = jornada + 1
print('\n-----------------------------------------------------------------------------------\n')
coste_total = calcular_coste(mejor_asignacion, matriz_tomas_actores)
print('El coste final es:', coste_total)
print('\n##################################################################################')
```
